# Webscrapping to name Columns

The selenium driver was used to scout through the [NHANES](https://wwwn.cdc.gov/nchs/nhanes/search/default.aspx) and the dictionary of the name was store.

We shall now rename the columns to more human reable descriptions

```python
import ast

with open('.\\data\\list_of_names.txt', 'r') as f:
    contents = f.read()
    column_name = ast.literal_eval(contents)
```

```python
data = data.rename(columns = column_name)
data.head()
```
